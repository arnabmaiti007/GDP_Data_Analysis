## GDP Analysis:

```python
import pandas as pd
def get_recession_start():
    '''Returns the year and quarter of the recession start time as a 
    string value in a format such as 2005q3'''
    
    df = pd.read_excel('gdp_data.xls')
    columns = list(df.columns)
    n = 0
    for i in columns:
        df.rename(columns={i : n}, inplace = True)
        n = n + 1
    df = df.drop([3, 7], axis='columns')
    df = df.drop([0, 1, 2, 3, 4, 5, 6], axis='rows')
    df = df.reset_index()
    
    new_df = df[[4, 6]]
    new_df.rename(columns={4 : 'year_quarter', 6 : 'gdp'}, inplace=True)
    
    splt = new_df['year_quarter'].str.split('q', expand=True)
    new_df[['year', 'quarter']] = splt
    
    nd = new_df.copy()
    start = 2000
    before_re = None
    for i in nd.index:
        if int(nd.iloc[i]['year']) >= start:
            a = float(nd.iloc[i]['gdp'])
            b = float(nd.iloc[i+1]['gdp'])
            c = float(nd.iloc[i+2]['gdp'])
            if before_re == None:
                before_re = nd['gdp'].min()
            if i + 3 == len(nd.index):
                break
            if a > b and b > c:
                if a > before_re:
                    before_re = a
                    re_start = b

    output = nd[nd['gdp'] == re_start]['year_quarter']
    output1 = output.copy().reset_index().loc[0]['year_quarter']

    return output1

def get_recession_bottom():
    '''Returns the year and quarter of the recession bottom time as a 
    string value in a format such as 2005q3'''
    df = pd.read_excel('gdplev.xls')
    columns = list(df.columns)
    n = 0
    for i in columns:
        df.rename(columns={i : n}, inplace = True)
        n = n + 1
    df = df.drop([3, 7], axis='columns')
    df = df.drop([0, 1, 2, 3, 4, 5, 6], axis='rows')
    df = df.reset_index()
    
    new_df = df[[4, 6]]
    new_df.rename(columns={4 : 'year_quarter', 6 : 'gdp'}, inplace=True)
    
    splt = new_df['year_quarter'].str.split('q', expand=True)
    new_df[['year', 'quarter']] = splt
    
    nd = new_df.copy()
    start = 2000
    reces_low = None
    for i in nd.index:
        if int(nd.iloc[i]['year']) >= start:
            a = float(nd.iloc[i]['gdp'])
            b = float(nd.iloc[i+1]['gdp'])
            c = float(nd.iloc[i+2]['gdp'])
            if reces_low == None:
                reces_low = nd['gdp'].max()
            if i + 3 == len(nd.index):
                break
            if a > b and b > c:
                if c < reces_low:
                    reces_low = c

    output = nd[nd['gdp'] == reces_low]['year_quarter']
    output1 = output.copy().reset_index().loc[0]['year_quarter']

    return output1
```

```python
recession_start = get_recession_start()
recession_bottom = get_recession_bottom()

print("The start of recession is - Year: {}, Quarter: {}".format(recession_start[:4], recession_start[4:]))
print("The Lowest GDP in recession is at - Year: {}, Quarter: {}".format(recession_bottom[:4], recession_bottom[4:]))
```
